## **Program Menghitung BMI**

```python
#Mengubah kolom tinggi badan dalam satuan cm ke m

df['tinggi_badan (m)'] = df['tinggi_badan (cm)'] / 100
```

```python
#Menghapus kolom "tinggi_badan (cm)"
#Karena sudah tidak digunakan lagi dan dapat mengganggu proses analisis lebih lanjut

df = df.drop('tinggi_badan (cm)', axis=1)
df
```

```python
#Function untuk menghitung BMI

def calculate_bmi(dataframe, weight_col, height_col):
    """
    Menghitung BMI (Indeks Massa Tubuh) dan menambahkannya sebagai kolom baru ke dalam DataFrame.

    Parameters:
    - dataframe: DataFrame yang ingin diubah.
    - weight_col: Nama kolom berat badan dalam DataFrame.
    - height_col: Nama kolom tinggi badan dalam DataFrame.

    Returns:
    - DataFrame asli dengan kolom BMI tambahan.
    """
    # Membuat salinan DataFrame agar tidak memodifikasi DataFrame asli.
    df = dataframe.copy()

    # Menghitung BMI dan menambahkannya sebagai kolom baru.
    df['BMI'] = df[weight_col] / (df[height_col] ** 2)

    return df
```

```python
#Memanggil dan menjalankan function yang sudah dideklarasikan

df_with_bmi = calculate_bmi(df, 'berat_badan (kg)', 'tinggi_badan (m)')

df_BMI = pd.DataFrame(df_with_bmi)
df_BMI
```

# **Program Memberikan Status BMI**

```python
#Function untuk memberikan tambahan status BMI pada dataframe

def add_bmi_status(dataframe):
    """
    Menambahkan kolom 'BMI Status' ke dalam DataFrame berdasarkan nilai BMI.

    Parameters:
    - dataframe: DataFrame yang ingin diubah.

    Returns:
    - DataFrame asli dengan kolom 'BMI Status' tambahan.
    """
    # Membuat salinan DataFrame agar tidak memodifikasi DataFrame asli.
    df = dataframe.copy()

    # Menentukan status BMI berdasarkan kategori yang Anda inginkan.
    def bmi_status(bmi):
        if bmi < 18.5:
            return 'Underweight'
        elif 18.5 <= bmi < 24.9:
            return 'Normal'
        elif 25 <= bmi < 29.9:
            return 'Overweight'
        else:
            return 'Obesity'

    # Menambahkan kolom 'BMI Status' dengan memanggil fungsi bmi_status.
    df['BMI Status'] = df['BMI'].apply(bmi_status)

    return df
```

```python
#Memanggil dan menjalankan function penambahan status BMI yang sudah dideklarasikan

df_BMI_status = add_bmi_status(df_BMI)

df_BMI_New = pd.DataFrame(df_BMI_status)
df_BMI_New
```
